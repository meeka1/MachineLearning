### Support Vector Machine (Classification model)

Note: line with higher margin is better

SVM draws a hyper plane in N-dimensional surface such that it maximizes margin margin between classification groups

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print(dir(iris))
```

```python
(iris.feature_names) #sepal - чашелистик, petal - лепесток
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df_target = pd.DataFrame(iris.target)
```

```python
df_target
```

```python
iris.target_names # setosa: 0, versicolor: 1, virginica:2 
```

```python
pd.concat([df, df_target], axis=1)
```

```python
df['target'] = df_target #adding target column to the dataframe
```

```python
df
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
# separating each flower in different dataframe
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
df0.head()
```

```python
plt.xlabel('sepal length (cm)', fontsize=16)
plt.ylabel('sepal width (cm)', fontsize=16)
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color="blue", marker="+", label="setosa")
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color="red", marker="*", label="versicolor")
plt.legend()
plt.show()
```

```python
plt.xlabel('petal length (cm)', fontsize=16)
plt.ylabel('petal width (cm)', fontsize=16)
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color="blue", marker="+", label="setosa")
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color="red", marker="*", label="versicolor")
plt.legend()
plt.show()
```

#### Train the model using SVM

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target', 'flower_name'], axis=1)
```

```python
X
```

```python
y = df.target
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train.shape, y_train.shape
```

```python
from sklearn.svm import SVC # support vector classifier
svc_model = SVC(C=1, kernel="rbf") # changing Regularization (C)
```

```python
svc_model.fit(X_train, y_train) #train the model
```

```python
svc_model.score(X_test, y_test) # calculates the accuracy
```

```python
import pickle
```

```python
with open('svc_model.pkl', 'wb') as f:
    pickle.dump(svc_model, f)
```

```python
with open('svc_model.pkl', 'rb') as f:
    mp = pickle.load(f)
```

```python
mp.predict([[5.1, 3.5, 1.4, 0.2]]) # array([0]) means setosa
```
