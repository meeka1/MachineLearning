```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
```

```python
df.rename(columns={'salary_more_then_100k':'salary_more_than_100k'}, inplace=True)
```

```python
df
```

```python
inputs = df.drop('salary_more_than_100k',axis=1)
target = df['salary_more_than_100k']
```

```python
pd.DataFrame(target)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_name'] = LabelEncoder().fit_transform(inputs['company'])
inputs['job_name'] = LabelEncoder().fit_transform(inputs['job'])
inputs['degree_name'] = LabelEncoder().fit_transform(inputs['degree'])
```

```python
inputs
```

```python
encoded_df = inputs.drop(['company', 'job', 'degree'], axis=1)
```

```python
encoded_df
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
X = encoded_df
y = target
```

```python
model.fit(X, y)
```

```python
model.score(X,y)
```

```python
model.predict([[2,2,1]])
```
