```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
```

```python
# Statistical description about patients who had heart disease
df[df['target']==1].describe()
```

```python
# Splitting the features and the target

X = df.drop('target', axis=1)
```

```python
y = df['target']
```

```python
X
```

### Training the model

```python
from sklearn.model_selection import train_test_split
```

```python
(X_train, X_test, y_train, y_test) = train_test_split(X, y, test_size=0.15)
```

```python
len(X_train), len(X_test)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### Train with Logistic Regression Classification model

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(C=10)
```

```python
log_model.fit(X_train, y_train)
```

```python
log_model.score(X_test, y_test)
```

```python
log_model.score(X_train, y_train)
```

### Building a confusion matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_predicted = log_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predicted)
```

```python
cm
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True)
plt.xlabel('actual')
plt.ylabel('predicted')
```

### Saving the trained model

```python
import joblib
```

```python
joblib.dump(log_model, 'log_model')
```
